```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json(r"D:\Downloads\My code\NASA-Asteroid-Classifier\data\raw_data.json")
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=df[["absolute_magnitude", "velocity", "diameter_min", 
                      "diameter_max", "miss_distance_au"]])
plt.title("Feature Distribution and Outliers")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = ["absolute_magnitude", "velocity", "diameter_min", "diameter_max", "miss_distance_au"]

correlation = df[numeric_cols].corr()

sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```
